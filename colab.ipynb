### Cleaning Data

Data Analyst hanya berkerja pada tabel order untuk menjawab pertanyaan pertama

```python
order_df.duplicated().sum()
```

```python
order_df['order_purchase_timestamp'] = pd.to_datetime(order_df['order_purchase_timestamp'])
order_df['purchase_month'] = order_df['order_purchase_timestamp'].dt.month

def classify_holiday(month):
    if month in [1, 4, 10, 12]:  # January, April, October, December
        return 'New Year / Easter / Halloween / Christmas'
    elif month in [5, 6, 7]:  # May, June, July (Eid Mubarak)
        return 'Eid Mubarak'
    elif month == 11:  # November (Thanksgiving)
        return 'Thanksgiving'
    elif month in [1, 2]:  # January, February (Chinese New Year)
        return 'Chinese New Year'
    else:
        return 'No Holiday'

order_df['holiday_name'] = order_df['purchase_month'].apply(classify_holiday)

order_df.head()
```

Data Analyst hanya berkerja pada tabel order_items dan order_reviews untuk menjawab pertanyaan kedua

```python
order_item_df.duplicated().sum()
```

```python
order_review_df.duplicated().sum()
```

```python
order_item_review_df = pd.merge(order_item_df, order_review_df, on="order_id", how="inner")

order_item_review_df.head()
```

```python
sales_by_holiday = order_df.groupby('holiday_name')['order_id'].count().reset_index()
sales_by_holiday_sorted = sales_by_holiday.sort_values(by='order_id', ascending=False)

print(sales_by_holiday_sorted)
```

### Explore table order_item_review

```python
order_item_review_df.sample(5)
```

```python
product_mean_price = order_item_review_df.groupby('product_id')['price'].mean().reset_index()

# mencari mean dari harga produk
mean_price = product_mean_price['price'].mean()

# filter produk yang memiliki harga dibawah rata-rata
below_mean_products = product_mean_price[product_mean_price['price'] < mean_price]

# kalkulasi rating untuk produk dengan mean dibawah rata-rata
product_mean_rating = order_item_review_df[order_item_review_df['product_id'].isin(below_mean_products['product_id'])].groupby('product_id')['review_score'].mean().reset_index()

# merge keduanya
result_df = pd.merge(product_mean_rating, below_mean_products, on="product_id", how="inner")

result_df.describe()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(sales_by_holiday_sorted['holiday_name'], sales_by_holiday_sorted['order_id'], color='skyblue')
plt.xlabel('Holiday Name')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Holiday')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

### Pertanyaan 2:

```python
result_df['price_quartile'] = pd.qcut(result_df['price'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x='review_score', y='price_quartile', data=result_df, color='skyblue', alpha=0.5)
plt.xlabel('Average Review Score')
plt.ylabel('Price Quartile')
plt.title('Distribution of Prices and Average Review Scores')
plt.tight_layout()

# Display the plot
plt.show()
```
